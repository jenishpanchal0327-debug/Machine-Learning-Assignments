```python
import numpy as np
import scipy.stats as stats

# Given data
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])

# Sample size
n = len(data)

# Sample mean
mean = np.mean(data)

# Sample standard deviation (n-1)
std_sample = np.std(data, ddof=1)

mean, std_sample, n
```

```python
# 99% confidence level
alpha = 0.01
df = n - 1

# t critical value
t_critical = stats.t.ppf(1 - alpha/2, df)

# Margin of error
margin_error_t = t_critical * (std_sample / np.sqrt(n))

# Confidence Interval
ci_t = (mean - margin_error_t, mean + margin_error_t)

ci_t
```

```python
# Given population standard deviation
sigma = 0.2

# z critical value
z_critical = stats.norm.ppf(1 - alpha/2)

# Margin of error
margin_error_z = z_critical * (sigma / np.sqrt(n))

# Confidence Interval
ci_z = (mean - margin_error_z, mean + margin_error_z)

ci_z
```

The sample size is small (n = 15) and the population standard deviation is unknown.
Therefore, the t-distribution is used to construct a 99% confidence interval for the mean durability of print-heads.
The interval is calculated using the sample mean and sample standard deviation.

When the population standard deviation is known (σ = 0.2), the z-distribution is appropriate.
A 99% confidence interval for the mean durability is constructed using the normal distribution.
