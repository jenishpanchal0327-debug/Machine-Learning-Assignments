# Hypothesis Testing

This notebook contains solutions for:
1. Chi-Square Test
2. One-Sample Z-Test

## Task 1: Chi-Square Test of Independence

```python
import numpy as np
from scipy.stats import chi2_contingency, chi2
```

```python
data = np.array([
    [40, 30, 10],  # Smart Thermostat
    [35, 25, 20]   # Smart Light
])

chi2_stat, p_value, dof, expected = chi2_contingency(data)
chi2_stat, p_value, dof, expected
```

```python
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, dof)
critical_value
```

**Conclusion:** If Chi-square statistic > Critical value, reject H₀.
There is a relationship between device type and customer satisfaction.

## Task 2: One-Sample Z-Test

```python
from scipy.stats import norm
import numpy as np
```

```python
X = 600
n = 25
sample_mean = 4100
sigma = 125
alpha = 0.05

mu = 1000 + 5 * X
z_stat = (sample_mean - mu) / (sigma / np.sqrt(n))
z_stat
```

```python
z_critical = norm.ppf(1 - alpha)
z_critical
```

**Conclusion:** If Z statistic > Z critical value, reject H₀.
Weekly operating costs are higher than the model predicts.
