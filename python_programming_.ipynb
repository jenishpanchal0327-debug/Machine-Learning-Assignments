```python
# Exercise 1: Check whether a number is prime

num = int(input("Enter a number: "))

if num <= 1:
    print(num, "is not a prime number")
else:
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, "is a prime number")
    else:
        print(num, "is not a prime number")
```

```python
# Exercise 2: Product of Random Numbers

import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

print("First number:", num1)
print("Second number:", num2)

user_answer = int(input("Enter the product of these two numbers: "))

correct_answer = num1 * num2

if user_answer == correct_answer:
    print("Correct! Well done")
else:
    print("Incorrect ")
    print("Correct answer is:", correct_answer)
```

```python
# Exercise 3: Squares of EVEN numbers from 100 to 200

print("Squares of even numbers between 100 and 200:")

for num in range(100, 201):
    if num % 2 == 0:   # checking even numbers
        print(f"{num}^2 =", num ** 2)
```

```python
# Exercise 4: Word Counter

input_text = "This is a sample text. This text will be used to demonstrate the word counter."

words = input_text.split()
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Printing result
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

```python
# Exercise 5: Palindrome Check

def is_palindrome(text):
    cleaned_text = ""

    for char in text:
        if char.isalnum():   # keeps only letters and numbers
            cleaned_text += char.lower()

    return cleaned_text == cleaned_text[::-1]


# Example test
input_text = "racecar"
print(is_palindrome(input_text))  # Expected output: True
```
